```python
import geopandas as gpd

# Load the shapefile data
file_path = "../data/raw/trecs__basemap__utm__poly__v1__building_footprints.shp"
data = gpd.read_file(file_path)

# Display basic information
data.head()
```

```python
# Check for missing values in each column
data.isnull().sum()
```

```python
# Count duplicate rows
data.duplicated().sum()
```

```python
# Summary statistics for numeric columns
data.describe()
```

```python
# Quick plot of building footprints
data.plot(figsize=(10, 10), color='blue', edgecolor='black')
```
